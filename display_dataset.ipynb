```python
df_prep = pd.read_csv('Flat_Final_dataset_moscow_.csv')
X = df_prep[['Район Город','класс', 'Отделка', 'Зона', 'Площадь', 'Тип Комнатности']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
scaled_df = pd.DataFrame(scaled_X, columns=X.columns)
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=0.5)
model.fit_predict(scaled_df)
labels = model.labels_
scaled_df['Labels'] = labels
```

```python
display_df = pd.concat([scaled_df.drop(['Район Город','класс', 'Отделка', 'Зона', 'Площадь', 'Тип Комнатности'], axis=1), df], axis=1, join='inner')
```
